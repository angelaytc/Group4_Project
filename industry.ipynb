```python
import pandas as pd

def clean_columns(df):
  df.columns = [c.lower().replace(' ', '_') for c in df.columns]
  return df
```

```python
df = pd.read_excel('/content/greentasks.xlsx', 'Table_7', skiprows=2).pipe(clean_columns).iloc[16:-1]
df
```

```python
df = df.set_index('year').unstack().to_frame(name='pct_green_tasks').reset_index().rename(columns={'level_0': 'industry'})
```

```python
df
```

```python
import plotly.express as px
pd.options.plotting.backend = "plotly"

# qualitative, sequential, carto, cmocean, colorbrewer, cyclical, diverging
px.colors.sequential.swatches()
```

```python
df.plot(
    kind='line', 
    x='year', 
    y='pct_green_tasks', 
    color='industry', 
    color_discrete_sequence=px.colors.qualitative.Safe,
)
```

```python
max_hours_by_industry = (df
                         .set_index('year').unstack().to_frame(name='hours').reset_index().rename(columns={'level_0': 'industry'})
                         .sort_values(['hours'], ascending=[False])
                         .groupby(['industry'], as_index=False)
                         .first()
)
max_hours_by_industry.sort_values(['hours'], ascending=[False]).head(10)
```
