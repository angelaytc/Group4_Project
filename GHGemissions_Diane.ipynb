```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_rows', None)
```

```python
#fetch data
ghg_excel = '/content/atmosphericemissionsghgintensitybyindustrycorrected.xls'
df = pd.read_excel(ghg_excel)
df
```

```python
#select the specific sheet, remove the first 2 rows, set the Sector Summary as the index column

df = pd.read_excel(ghg_excel, sheet_name='GHG intensity', skiprows=3, index_col='Sector summary')
df.head()
```

```python
#specific coverage of the table: the main sector summary from rows A to V and column years= 2011-2018
ghg_df = df.iloc[0:20, 23:31]
ghg_df
```

```python
#List of Industries
ghg_df.index
```

```python
#number of sector/industries (rows) and years (columns)
ghg_df.shape
```

```python
# check overview of the elements and for the data type
ghg_df.info()
```

```python
#convert string (object) numeric values to float (int) data type
ghg_df = ghg_df.apply(lambda x : x.astype('float'))
ghg_df.dtypes
```

```python
# quick statistic summary of the data
ghg_df.describe()
```

```python
#find the mean value of GHG Emissions over the years (2011-2018) of each industry
ghg_df['Mean Value'] = ghg_df.mean(axis=1).sort_values(ascending=True)
ghg_df.sort_values(by=["Mean Value"], inplace=True, ascending=False)
ghg_df
```

```python
#catch the 10 highest GHG among the industries
max_ghg_df = ghg_df.max(axis=1).sort_values(ascending=False)
max_ghg_df.head(10).plot.bar()
```

```python
#catch the 10 lowest GHG among the industries
min_ghg_df = ghg_df.min(axis=1).sort_values(ascending=True)
min_ghg_df.head(10).plot.bar()
```

```python
#Use heatmap diagram showing which industry and year has highest emission
df_diagram = ghg_df.iloc[:, 0:8]

plt.figure(figsize=(10,10))
sns.heatmap(df_diagram, cmap="YlGnBu")
```

```python
#Transpose the dataframe for visualization
df_ghg_tanspose = df_diagram.transpose()
df_ghg_tanspose
```

```python
#Highest GHG Emissions
df_ghg_tanspose[["Electricity, gas, steam and air conditioning supply",
"Agriculture, forestry and fishing",
"Mining and quarrying",
"Water supply; sewerage, waste management and remediation activities",
"Transport and storage",
"Manufacturing",
"Construction",
"Wholesale and retail trade; repair of motor vehicles and motorcycles",
"Accommodation and food services",
"Public administration and defence; compulsory social security"]].plot(ylim=0, figsize=(25,15))
```

```python
#Lowest GHG Emissions
df_ghg_tanspose[["Financial and insurance activities",
"Real estate activities",
"Information and communication",
"Activities of households as employers; undifferentiated goods and services-producing activities of households for own use",
"Professional, scientific and technical activities",
"Education",
"Other service activities",
"Arts, entertainment and recreation",
"Administrative and support service activities",
"Human health and social work activities" ]].plot(ylim=0, figsize=(25,15))
```

```python
#Pie Chart of Highest GHG Emission per Industries (mean values)
slices_mean = [6.77, 4.21, 1.94, 1.55, 1.23, 0.53, 0.14, 0.11, 0.09, 0.07]
labels = ['Electricity, gas, steam and air conditioning supply',
    'Agriculture, forestry and fishing',
    'Mining and quarrying',
    'Water supply; sewerage, waste management and remediation activities',
    'Transport and storage',
    'Manufacturing',
    'Construction',
    'Wholesale and retail trade; repair of motor vehicles and motorcycles',
    'Accommodation and food services',
    'Public administration and defence; compulsory social security']
colors = ['blue','orange','green','red','purple','brown','pink','gray','yellow','cyan']
plt.figure(figsize=(10,10))
plt.pie(
    slices_mean,
    labels=labels,
    colors=colors,
    startangle=45,
    shadow= True,
    explode=(0.05,0,0,0,0,0,0,0.1,0.2,0.3),
    autopct='%1.1f%%'
)
plt.title('Industries with the Highest GHG Emission in UK (2011-2018)')
plt.show
```
