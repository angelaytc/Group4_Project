```python
import pandas as pd

def clean_columns(df):
  df.columns = [c.lower().replace(' ', '_') for c in df.columns]
  return df
```

```python
df_1 = pd.read_excel('/content/greentimesharesoc.xlsx', 'SOC_2000', skiprows=2).pipe(clean_columns)
df_2 = pd.read_excel('/content/greentimesharesoc.xlsx', 'SOC_2010', skiprows=2).pipe(clean_columns)

display(df_1.head(5))
display(df_2.head(5))
```

```python
# The majority of job categories (70%) from 1997 even to 2019 have 0% of working time spent on green related activities
# Of those job categories (the remaining 30%) that do some green related tasks, only 6% of job catergories spent more than a third
# of time doing green related tasks, this general pattern has held over 20 years.

# One interesting question would be to understand how many workers fall into each catergory
# so that we can then understand how involved the the entire UK workforce in green tasks

# One note of caution here is that the catergories may contain vastly different numbers of workers
# i.e. Chief executives and senior official posisitions are likely very limited compared to shelf stackers
# 

percentiles = [i/100 for i in range(0, 90, 10)] + [i/100 for i in range(90, 101, 2)]

display(df_1.describe(percentiles=percentiles).T.iloc[1: , :-1])
display(df_2.describe(percentiles=percentiles).T.iloc[1: , :-1])
```

```python
## We would need to do much more work to join these two datasets, as the SOC codes for 2000/2010 are not always directly comparable
df = (
    pd.merge(df_1, df_2, left_on='soc_2000_code', right_on='soc_2010_code', how='inner')
    .drop(columns=['soc_2000_code','soc_2010_code', 'soc_2000_description','2011_x','2019'])
    .set_index('soc_2010_description')
)
print(df.shape)
df.head()
```

```python
plot_df = df.reset_index().melt(id_vars=['soc_2010_description'], var_name='year', value_name='pct_green_tasks')
plot_df['year'] = pd.Categorical(values=plot_df['year'], categories=plot_df['year'].unique(), ordered=True)
plot_df.head(3)
```

```python
import plotly.express as px
pd.options.plotting.backend = "plotly"

# qualitative, sequential, carto, cmocean, colorbrewer, cyclical, diverging
px.colors.sequential.swatches()
```

# Notes:
There is an adjustment for changes in the occupation classification in 2011.

```python
plot_df.plot(
    kind='line', 
    x='year', 
    y='pct_green_tasks', 
    color='soc_2010_description', 
    color_discrete_sequence=px.colors.qualitative.Safe,
)
```

## Time spent doing green tasks by country
Proportion of hours worked spent on green tasks, whole economy by country of the UK

```python
df.plot(
    kind='line', 
    x='Year', 
    y='pct_green_tasks', 
    color='Country', 
    color_discrete_sequence=px.colors.qualitative.Safe,
)
```

England is estimated to have the greatest share of hours worked on green tasks (at 7.2%), slightly above the UK average. While estimates for Scotland, Wales and Northern Ireland are a little lower at around 7% in Scotland and Wales, and 6.4% in Northern Ireland.
